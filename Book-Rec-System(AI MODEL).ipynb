```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df = ratings_with_name.groupby('Title', as_index=False)['Rating'].count()
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Title', as_index=False)['Rating'].mean()
avg_rating_df = avg_rating_df.rename(columns={'Rating': 'Avg-Rating'})
avg_rating_df

```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Rating']>=250].sort_values('Avg-Rating',ascending=False).head(50)
popular_df
```

**COLLABORATIVE FILTERING BASED RECOMMENDATION SYSTEM**

```python
# Only taking ratings of experienced Users to make the system more intelligent (More than 200 books rated)
x=ratings_with_name.groupby('User-ID').count()['Rating'] > 200
experienced_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'].isin(experienced_users)]
```

```python
y = filtered_ratings.groupby('Title').count()['Rating']
threshold = 50  # You can change this value as needed
famous_books = y[y > threshold].index

```

```python
final_ratings = filtered_ratings[filtered_ratings['Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Title',columns='User-ID',values='Rating')
pt.fillna(0,inplace=True)
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt).shape
```

```python

# Compute cosine similarity correctly
similarity_scores = cosine_similarity(pt)  # ✅ This should give (num_books, num_books)

print("Corrected shape of similarity_scores:", similarity_scores.shape)  # Debugging check
```

```python
def recommend(Book_name):
    # Get the index of the book
    index_array = np.where(pt.index == Book_name)[0]

    if len(index_array) == 0:
        print("Error: Book not found in dataset.")
        return

    index = index_array[0]

    # Sort similar items based on similarity scores
    similar_items = sorted(
        list(enumerate(similarity_scores[index])),
        key=lambda x: x[1],
        reverse=True
    )[1:6]  #  Get only the top 5 similar books

    # Print recommended books
    print(f"Books similar to '{Book_name}':")
    for i in similar_items:
        print(pt.index[i[0]])
```

```python
recommend('The Joy Luck Club')
```
